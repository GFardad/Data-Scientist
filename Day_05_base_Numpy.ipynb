```python
import numpy as np
import time
import torch
```

```python
for x in range(10000):
    array_1m1 = torch.rand(x)
    array_1m2 = torch.rand(x)
    array_result = []
    t0 = time.time()
    for a, (i, j) in enumerate(zip(array_1m1, array_1m2)):
        array_result.append(i * j)
    t1 = time.time()
    timea = t1 - t0

    t0 = time.time()
    array_1m1.dot(array_1m2)
    timetake = (timea/(time.time() - t0))
    print(
        f" time a : {timea} \n time b : {time.time() - t0 }\n numpy is : {timetake} times faster"
    )
```
