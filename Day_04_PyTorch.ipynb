```python
x = torch.tensor(2.0, requires_grad=True)
y = 3 * x**2 + 5
print(f"y  {y} \n x : {x},\n gradian x : {x.grad} \n gradian y : {y.backward()}")
```

```python
import torch

# 1. تعریف تنسور ورودی (a) با requires_grad=True
# a = [2.0, 3.0]
a = torch.tensor([1.0], requires_grad=True)

# 2. عملیات: b = 2a (اینجا b خروجی یا y است)
b = a**2 * 2**2

# 3. تعریف تنسور وزن گرادیان (v)
# v = [0.1, 1000.0]
v = torch.tensor([1.0])

# 4. محاسبه گرادیان نهایی (مشتق)
# PyTorch مشتق db/da را محاسبه و در v ضرب می‌کند
b.backward(v)

# 5. پرینت گرادیان نهایی ذخیره‌شده در a.grad
print(f"گرادیان a (a.grad): {a.grad}")
print(f"a : {b}")
```
