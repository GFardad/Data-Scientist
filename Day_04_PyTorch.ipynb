```python


# 1. ایجاد آرایه‌ی NumPy ۲×۲ با dtype استاندارد float32 (برای محاسبات AI)
numpy_array = np.array([[1.0, 2.0], [3.0, 4.0]], dtype=np.float32)

# 2. تبدیل به Tensor
torch_tensor = torch.tensor(numpy_array)

# 3. پرینت مشخصات
print(f"NumPy Array Type: {type(numpy_array)}")
print(f"NumPy Array Shape: {numpy_array.shape}")
print("-" * 30)
print(f"PyTorch Tensor Type: {type(torch_tensor)}")
print(f"PyTorch Tensor Shape: {torch_tensor.shape}")
print(f"PyTorch Tensor Dtype: {torch_tensor.dtype}")
```
