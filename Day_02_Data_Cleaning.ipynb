```python
# ایجاد DataFrame با داده‌های گمشده و تکراری
data_cleaning = {
    "TransactionID": [101, 102, 103, 104, 105, 105],  # 105 تکراری است
    "Product": ["Laptop", "Monitor", "Mouse", None, "Keyboard", "Monitor"],
    "Price": [1200, 300, 50, 45, None, 300],
    "Quantity": [1, 2, None, 1, 1, 2],  # دارای NaN
}
df_clean = pd.DataFrame(data_cleaning)
# print(df_clean)
# printo()
df_clean = df_clean.drop_duplicates(subset=["TransactionID"])
print(df_clean)
```

```python
mean_price = df_clean["Price"].mean()
df_clean = df_clean.fillna({"Quantity": 1, "Price": df_clean["Price"].mean()}).dropna(
    subset=["Product"]
)
df_clean = df_clean.fillna(0)
print(f"{df_clean}\n Mean of Price is : {mean_price}")
df_clean.to_csv("df_clean.csv", index=False)
```
