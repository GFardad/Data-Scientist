```python
import pandas as pd
import numpy as np

df_eda = pd.read_csv("cleaned_data.csv")

print(df_eda)

product_summery = df_eda.groupby("Product")["Price"].agg(["mean", "sum"])
print(product_summery)
```

```python
import numpy as np

df_eda["Total"] = df_eda["Price"] * df_eda["Quantity"]


pivot_sumry = df_eda.pivot_table(
    index="Product", values=["Price", "Quantity", "Total"], aggfunc="sum"
)

pivot_sumry["Bulk"] = np.where(pivot_sumry["Quantity"] > 5, True, False)


print(pivot_sumry)
```

```python
pivot_final = pivot_sumry.pivot_table(
    index="Product",
    values=["Price", "Quantity", "Total"],
    aggfunc={"Price": "sum", "Quantity": "mean", "Total": "mean"},
)
cta = ["Quantity", "Total", "Price"]
pivot_final[cta] = pivot_final[cta].apply(lambda x: x.round(0).astype(int))
print(pivot_final)
```
