## تحلیل داده‌ی اولیه

این جدول اولیه، داده‌های فرضی برای تمرین Indexing و Selection را نشان می‌دهد. Age و Income متغیرهای عددی ما هستند.

```python
import pandas as pd
def prin ():
    print(" \n ","-"*10 ," \n " )

data = {
    "Name": ["Ali", "Sara", "Reza", "Mina", "Vahid"],
    "Age": [22, 35, 28, 41, 30],
    "Income_Mln_Toman": [30, 80, 55, 120, 45],
}

df = pd.DataFrame(data)
print(df)
prin()
```

```python
# 1. Selection by Position (iloc)
# انتخاب سطرهای 0 تا 2 (سطر 3 شامل نمی شود) و ستون 1 (Age)
pos_select = df.iloc[0:3, 1]
print("Selection using iloc (Position):\n", pos_select)

# 2. Selection by Label (loc)
# انتخاب سطرهایی که نامشان 'Sara' است و ستون Income_Mln_Toman
label_select = df.loc[df["Name"] == "Sara", "Income_Mln_Toman"]
print("\nSelection using loc (Label):\n", label_select)
```

```python
df_filter = df[df["Age"] > 30 & df[Income_Mln_Toman] > 50]
```
